```python
chaines=["HILTON", "HOLIDAY", "WESTERN", "RENAISSANCE", "NOVOTEL", "CAMPANILE", "ETAP", 
"MERCURE", "COMFORT", "KYRIAD", "HYATT", "TIMHOTEL", "IBIS", "F1", "PREMIERE", 
"PREMIÈRE", "CLASSICS", "SPLENDID", "FASTHOTEL", "NEWHOTEL","BRIT", "BALLADINS",
"RADISSON", "SOFITEL", "INTERCONTINENTAL", "B&B", "MARRIOTT", "CROWNE", "PULLMAN"]
```

```python
chaines=[str.lower(x) for x in chaines]
```

```python
hotels["name"]=[str.lower(x) for x in hotels['name']]
```

```python
hotels["chaine"]=[1 if any(x in Z for x in chaines) else 0 for Z in hotels["name"]]
```
