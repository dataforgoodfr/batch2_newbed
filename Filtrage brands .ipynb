```python
hotels=pd.read_csv("TripAdvisor_hotels_with_serv_and_brands")
hotels
```

```python
hotels.columns
```

```python
hotels=hotels.drop(['extended_address','Unnamed: 0',"id",'non-smoking_hotel', 'non-smoking_rooms','minibar','bar/lounge', 
                    'spa','fitness_center_with_gym_/_workout_room', 'pool', 'suites','hot_tub',
       'smoking_rooms_available','tennis_court', 'beachfront',
       'golf_course', 'casino_and_gambling', 'ski-in_/_ski-out', 'country'],axis=1)
```

```python
hotels
```

```python
hotels=hotels[hotels["star_rating"]<4]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.hist(hotels["bubble_rating"].dropna())
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.hist(hotels["review_count"].dropna(), bins=100)
plt.show()
```

```python
hotels["review_count"].mean()
```

```python
hotels["bubble_rating"].mean()
```

```python
hotels["rating_std"]=hotels["bubble_rating"]*hotels["review_count"]/((hotels["bubble_rating"]*hotels["review_count"]).mean())
```

```python
hotels["rating_std2"]=0.5*(hotels["bubble_rating"]/(hotels["bubble_rating"].mean())+hotels["review_count"]/hotels["review_count"].mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.hist(hotels["rating_std"].dropna(), range=[0,10], bins=100)
plt.show()
```

```python
plt.hist(hotels["rating_std2"].dropna(), range=[0,10], bins=100)
plt.show()
```

```python
#On peut ne garder que les 75% d'hotels les moins bien notÃ©s
hotels["rating_std2"].dropna().quantile(.75)
```

```python
hotels_low=hotels[hotels["rating_std2"]<1.11]
```
