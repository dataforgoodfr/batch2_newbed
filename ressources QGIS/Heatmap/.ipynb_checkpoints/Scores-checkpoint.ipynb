```python
import pandas as pd 
```

```python
Communes=pd.read_csv("Comm_singa.csv")
```

```python
Communes
```

```python
from numpy import mean
Moyennes=[mean(Communes[x]) for x in Communes.columns[1:] ]
```

```python
Moyennes=[mean(Communes[x]) for x in Communes.columns[1:] ]
Communes["score"]=(4*(Communes["ACC_ASILE,C,1"]/Moyennes[Communes.columns.get_loc("ACC_ASILE,C,1")-1]+
                    Communes["CHRS,C,2"]/Moyennes[Communes.columns.get_loc("CHRS,C,2")-1]+
                    Communes["Asso_singa,N,10,0"]/Moyennes[Communes.columns.get_loc("Asso_singa,N,10,0")-1])+
                3*(Communes["ENF_HEBER,C,2"]/Moyennes[Communes.columns.get_loc("ENF_HEBER,C,2")-1]+
                Communes["ACT_EDUC,C,1"]/Moyennes[Communes.columns.get_loc("ACT_EDUC,C,1")-1]+
                Communes["TRAV_PROT,C,2"]/Moyennes[Communes.columns.get_loc("TRAV_PROT,C,2")-1]+
                Communes["Volo_singa,N,10,0"]/Moyennes[Communes.columns.get_loc("Volo_singa,N,10,0")-1])+
                2*(Communes["ETA_PSY,C,2"]/Moyennes[Communes.columns.get_loc("ETA_PSY,C,2")-1]+
                Communes["URGENCES,C,1"]/Moyennes[Communes.columns.get_loc("URGENCES,C,1")-1]+
                  Communes["MATERNITE,C,1"]/Moyennes[Communes.columns.get_loc("MATERNITE,C,1")-1])+
                Communes["AUTRES_SOC,C,2"]/Moyennes[Communes.columns.get_loc("AUTRES_SOC,C,2")-1]+
                Communes["HEB_PROV,C,1"]/Moyennes[Communes.columns.get_loc("HEB_PROV,C,1")-1])/(4*3+3*4+2*3+2)
```

```python
Communes["score"].to_csv("Comm_score.csv")
```
