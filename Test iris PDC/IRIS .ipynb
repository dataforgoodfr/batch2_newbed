```python
import shapefile as shp
import shapely
```

```python
IRIS_sf=shp.Reader("CONTOURS-IRIS_2-0_SHP_LAMB93_D059-2014/CONTOURS-IRIS_D059.shp")
```

```python
IRIS_shapes=IRIS_sf.shapes()
IRIS_points=[q.points for q in IRIS_shapes]
```

```python
from shapely.geometry import Polygon
polygons=[Polygon(q) for q in IRIS_points]
```

```python
Hotels_pdc=shp.Reader("PDC_shp.shp")
```

```python
Hotels_shapes=Hotels_pdc.shapes()
```

```python
from shapely.geometry import Point
Hotels_coords=[q.points[0] for q in Hotels_shapes]
Hotels=[Point(q.points[0]) for q in Hotels_shapes]
```

```python
from rtree import index 
idx=index.Index()
c=-1
for q in IRIS_shapes:
    c+=1
    idx.insert(c, q.bbox)
```

```python
matches=[]
for i in range(len(Hotels)):
    temp=None
    print("Point", i)
    for j in idx.intersection(Hotels_coords[i]):
        if Hotels[i].within(polygons[j]):
            print( "Match found!", j)
            temp=j
            break
    matches.append(temp)
```
